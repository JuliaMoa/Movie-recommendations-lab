```python
import pandas as pd
from collections import Counter

df = pd.read_csv("movies_clean.csv")

# Dela upp filmer i två grupper
high = df[df["rating"] > 4.0]
low = df[df["rating"] < 3.0]

# Funktion för att räkna ord
def word_freq(series):
    words = " ".join(series).split()
    return Counter(words)

# Räkna ord i båda grupperna
high_freq = word_freq(high["tags"])
low_freq = word_freq(low["tags"])

# Visa de 20 vanligaste orden
print("Vanligaste ord i hög-rated filmer:")
print(high_freq.most_common(20))

print("\nVanligaste ord i låg-rated filmer:")
print(low_freq.most_common(20))
```
